Given a signed 32-bit integer x, return x with its digits reversed. If reversing x causes the value to go outside the signed 32-bit integer range [-231, 231 - 1], then return 0.

Assume the environment does not allow you to store 64-bit integers (signed or unsigned).

Input: x = 123

Output: 321

Input: x = -123

Output: -321

Input: x = 120

Output: 21

```python
def reverse(x):

  if x < 0 : # negative number convert to string and remove - from last and reverse and convert to int Again and add -  
      x = -int(str(x)[len(str(x))-1:0:-1])
  elif x > 0:# convert to string and reverse and convert to int again 
      x = int(str(x)[::-1])

  if (x > (-2**31) ) and (x < (2**31) ): # return only number in range 32bit 
    return x
  else :
    return 0 
```
